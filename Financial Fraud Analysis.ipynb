```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('synthetic_fraud_dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
numeric_cols= df.select_dtypes(include=['int64','object']).columns
cat_cols= df.select_dtypes(include=['float64']).columns

print('Numerical columns are:', numeric_cols.tolist())
print('Categorical columns are:', cat_cols.tolist())
```

```python
sns.countplot(data=df,x='is_fraud', palette='muted', hue='is_fraud', legend=False)
plt.title('Fraud vs non-Fraud')
plt.show
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python

ax=sns.countplot(data=df, x='transaction_type',hue='transaction_type',  palette='muted')
ax.set_axisbelow(True)
ax.grid(linestyle='--', alpha=0.7, axis='y')
plt.title('Transaction Types')
plt.show
```

```python

sns.countplot(data=df, y= 'merchant_category',hue='merchant_category' ,order=df['merchant_category'].value_counts().index, palette='muted')

plt.title('Merchant Category Distribution')
plt.show
```

```python
sns.barplot(data=df, x='transaction_type', y=df['is_fraud'],hue='transaction_type',  palette='muted')

plt.title('Fraud by Transaction Type')
```
